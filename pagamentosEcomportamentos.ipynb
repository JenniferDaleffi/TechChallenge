***Uso de parcelamento***


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# 1. Filtrar apenas cartão de crédito
df_credit = df[df["payment_type"] == "credit_card"].copy()

# 2. Agrupar por parcelas
df_plot = df_credit.groupby("payment_installments").size().reset_index(name="quantidade")

# Remover valores nulos ou 0
df_plot = df_plot[df_plot["payment_installments"].notna()]
df_plot = df_plot[df_plot["payment_installments"] > 0]

# Ordenar
df_plot = df_plot.sort_values("payment_installments")

# 3. Criar gráfico
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(
    df_plot["payment_installments"].astype(int).astype(str) + "x",
    df_plot["quantidade"]
)

# 4. Adicionar valores nas barras
for bar in bars:
    width = bar.get_width()
    ax.text(
        width,
        bar.get_y() + bar.get_height()/2,
        f'{int(width):,}'.replace(',', '.'),
        va='center',
        ha='left',
        fontsize=10
    )

# 5. Ajustes visuais
ax.set_xlabel("Quantidade de compras", fontsize=11)
ax.set_ylabel("Número de parcelas", fontsize=11)
ax.set_title("Quantidade de compras por número de parcelas - Cartão de crédito", fontsize=13)

bars = ax.barh(
    df_plot["payment_installments"].astype(int).astype(str) + "x",
    df_plot["quantidade"],
    color="#F28E2B"
)

# Formatação de milhar
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.')))

# Grid leve
ax.grid(axis='x', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

***Perfil financeiro***

```python
def classificar_cliente(row):
    if row['payment_type'] == 'boleto':
        return 'Conservador'
    elif row['payment_value'] > 200:
        return 'Alavancado'
    else:
        return 'Moderado'

df['perfil_cliente'] = df.apply(classificar_cliente, axis=1)

perfil = df['perfil_cliente'].value_counts()

plt.figure()
perfil.plot(kind='bar')
plt.title("Perfil Financeiro dos Clientes")
plt.show()
```

```python
df_pagamento = df.groupby(
    ["perfil_cliente", "payment_type"]
).size().reset_index(name="quantidade")
```

```python
import matplotlib.pyplot as plt

# Pivot
pivot = df_pagamento.pivot(
    index="perfil_cliente",
    columns="payment_type",
    values="quantidade"
).fillna(0)

# Plot
ax = pivot.plot(kind="bar", figsize=(8,6))

plt.title("Método de pagamento por perfil do cliente")
plt.xlabel("Perfil do cliente")
plt.ylabel("Quantidade de compras")
plt.legend(title="Método de pagamento")

# Adicionar valores nas barras
for p in ax.patches:
    height = p.get_height()

    if height > 0:
        ax.annotate(
            f'{int(height):,}'.replace(',', '.'),
            (p.get_x() + p.get_width() / 2, height),
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.tight_layout()
plt.show()
```

***Ticket médio por perfil***

```python
ticket = df.groupby('perfil_cliente')['payment_value'].mean()

plt.figure()
ticket.plot(kind='bar')
plt.title("Ticket Médio por Perfil")
plt.ylabel("Valor Médio")
plt.show()
```
